```python
from __future__ import print_function
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import numpy as np
import torchvision.models as models
import os, sys, json
from model import TrackNet
from utils.get_dataloader import get_dataloader
import cv2
from PIL import Image
import datetime
```

```python

cuda0 = torch.device('cuda:0')
model = TrackNet().to(cuda0)

model.load_state_dict(torch.load('weight/epoch_103_2580'))
model.eval()


def open_image(path, shape):
    image = Image.open(path).convert("RGB")
    return np.asarray(image.resize(shape))
```

```python
image3 = open_image('../data/DJI_0015/game_img_10695.jpg', (640,360))
image2 = open_image('../data/DJI_0015/game_img_10696.jpg', (640,360))
image1 = open_image('../data/DJI_0015/game_img_10697.jpg', (640,360))
image = np.dstack([image1, image2, image3])
print(image.shape)
image = image.transpose((2, 0, 1))
image = image[np.newaxis, :]
print(image.shape)
image = torch.from_numpy(image).float().to(cuda0)
print(image.shape)
output = model(image) * 255
```

```python
import matplotlib.pyplot as plt
print(output.shape)
output_numpy = output.squeeze().cpu().detach().numpy()
plt.imshow(output_numpy, cmap = "gray")
plt.show()
plt.imshow(image1)
plt.show()
```

```python
cap = cv2.VideoCapture('../data/DJI_0015.MP4')
count = -1
image1 = image2 = image3 = None
while True:
    count += 1
    ret, frame = cap.read()
    if ret:
        image = cv2.resize(frame, (640, 360))
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        if count == 0:
            image1 = image
        elif count == 1:
            image2 = image
        elif count == 2:
            image3 = image
        else:
            image1, image2 = image2, image3
            image3 = image
            image = np.dstack([image1, image2, image3])
            image = image.transpose((2, 0, 1))
            image = image[np.newaxis, :]
            image = torch.from_numpy(image).float().to(cuda0)
            output = model(image)
            output_numpy = output.squeeze().cpu().detach().numpy()
            #output_numpy = cv2.cvtColor(output_numpy, cv2.COLOR_RGB2GRAY)
            #output_numpy = cv2.cvtColor(output_numpy, cv2.COLOR_GRAY2RGB)
    #         print(output_numpy.shape)
    #         print(image_numpy.shape)
    #         plt.imshow(output_numpy)
    #         plt.show()
    #         plt.imshow(image_numpy)
    #         plt.show()

            write_image = cv2.addWeighted(src1=cv2.cvtColor(image3, cv2.COLOR_RGB2GRAY),alpha=0.4,src2=output_numpy,beta=0.6,gamma=0, dtype=cv2.CV_32F)
            #write_image = cv2.bitwise_and(src1=image_numpy, src2=output_numpy)
            cv2.imwrite('output/{:05d}.jpg'.format(count), write_image)
    if count % 100 == 0:
        print(datetime.datetime.now(), count)
```
