```python
from __future__ import print_function
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torchvision
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import numpy as np
import torchvision.models as models
import os, sys, json
from model import TrackNet
from utils.get_dataloader import get_dataloader
import cv2
```

```python

cuda0 = torch.device('cuda:0')
model = TrackNet().to(cuda0)

model.load_state_dict(torch.load('weight/epoch_0_1299'))
model.eval()
```

```python
image_numpy = cv2.imread('../data/tracking_data/vott-json-export/DJI_0013_30.MP4#t=14.833333.jpg')
image_numpy = cv2.resize(image_numpy, (640, 360))
image = torch.from_numpy(image_numpy).float().to(cuda0)
image = image.transpose(2, 0)
image = image[np.newaxis, :]
print(image.shape)
output = model(image)
```

```python
import matplotlib.pyplot as plt
print(output.shape)
output_numpy = output.squeeze().cpu().detach().numpy()
plt.imshow(output_numpy.T * 255)
plt.show()
plt.imshow(image_numpy)
plt.show()
```
